```python
features = [
    'buildingCount', 'totalHeight', 'avgHeight', 'basePerimeterTotal',
    'basePerimeterAvg', 'compactness', 'baseAreaTotal', 'totalArea',
    'avgBuildingArea', 'FAR', 'coverageRatio',
    'roadDensity', 'intersectionDensity', 'poiDensity', 'poiDiversity',
    'FVC', 'permeableRatio', 'maxHeight', 'minHeight', 'heightRange', 'heightIndex',
    'heightDensity', 'heightVariety', 'cornerCountTotal', 'basePerimeterMax',
    'basePerimeterMin', 'shapeComplexity', 'areaVariance', 'parcelArea',
    'largestPatchIndex', 'shape3DIndex', 'balanceIndex', 'cornerCountAvg',
    'buildingProximity', 'buildingMinDist', 'buildingMaxDist',
    'buildingDistAvg', 'buildingDistVar', 'SVF', 'streetRatio'
]
```

```python
# Extract the selected feature subset
df_feat = grid500[features].copy()

# Identify skewed features for transformation
skewness = df_feat.skew()
skewed_cols = skewness[skewness > 1].index.tolist()  # Skewness > 1 = strongly right-skewed
print(skewed_cols)
```

```python
len(skewed_cols)
```

```python
# Create copy for transformation
grid500_transformed = grid500.copy()

# Define a small epsilon shift to ensure positivity
epsilon = 1e-6  # or you can use slightly larger like 1e-3

for col in skewed_cols:
    min_val = grid500_transformed[col].min()
    if min_val <= 0:
        shift = abs(min_val) + epsilon
        print(f" Shifting {col} by +{shift:.2e} to ensure positivity before log1p.")
        grid500_transformed[f'log_{col}'] = np.log1p(grid500_transformed[col] + shift)
    else:
        grid500_transformed[f'log_{col}'] = np.log1p(grid500_transformed[col])

grid500_transformed.describe()
```

```python
log_features = [f'log_{col}' for col in skewed_cols]
non_log_features = [col for col in features if col not in skewed_cols]

features_transformed = log_features + non_log_features

print(len(features_transformed))
features_transformed
```

```python
for col in skewed_cols:
    log_col = f'log_{col}'
    if log_col in grid500_transformed.columns:
        corr_before = grid500_transformed[col].corr(grid500_transformed['log_VIT202311'])
        corr_after = grid500_transformed[log_col].corr(grid500_transformed['log_VIT202311'])
        print(f"{col}: corr before = {corr_before:.3f}, after log = {corr_after:.3f}")
```

### Conclusion: Log-transforming predictors did not improve model R and MSE
- The XGBoost model (log_VIT202311) achieved the same R (0.6860) and MSE (1.14) across all tested versions:
    - Using original features
    - Using log-transformed features
    - Using both original + log-transformed features
- Although log transformation improved the correlation between individual features and the target variable, it did not enhance model performance.
- This is because XGBoost is a non-linear, tree-based model that can naturally handle skewed or non-normal feature distributions without needing transformations.
- Therefore, for this task and model type, log-transforming the predictors is not necessary from a predictive performance standpoint.

```python
# Set global random seed for reproducibility
SEED = 42
np.random.seed(SEED)

# Ensure output directories exist
os.makedirs("../output/models", exist_ok=True)
os.makedirs("../output/param_logs", exist_ok=True)

# ========================= XGBoost Training Loop =========================
targets = ['log_NTL2023_focal', 'log_VIT202311', 'UHIDAY2020_07_inv']

for target in targets:
    print(f"\n===== Training model for target: {target} =====")

    # Filter rows with non-missing values for target and features
    df_target = grid500.copy()
    
    # Split features and target
    X = df_target[features]
    y = df_target[target]

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=SEED
    )

    print(f"Training samples: {X_train.shape[0]}, Testing samples: {X_test.shape[0]}")

    # Define base XGBoost model
    xgb_model = xgb.XGBRegressor(
        objective='reg:squarederror',
        random_state=SEED,
        n_jobs=-1
    )

    # Randomized hyperparameter search
    param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.03, 0.05, 0.07, 0.1],
    'subsample': [0.7, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.9, 1.0],
    'gamma': [0, 0.1, 0.3],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [1, 1.5, 2],
    'min_child_weight': [1, 3, 5],
    }
    
    random_search = RandomizedSearchCV(
        estimator=xgb_model,
        param_distributions=param_dist,
        n_iter=200,
        cv=5,
        scoring='neg_mean_squared_error',
        verbose=1,
        random_state=SEED
    )

    # Fit model
    random_search.fit(X_train, y_train)

    # Retrieve best estimator
    best_model = random_search.best_estimator_

    # Save trained model
    model_path = f"../output/models/{target}_xgb_model.pkl"
    joblib.dump(best_model, model_path)
    print(f" Saved best model to: {model_path}")

    # Save best hyperparameters
    best_params = random_search.best_params_
    param_path = f"../output/param_logs/{target}_best_params.json"
    with open(param_path, 'w') as f:
        json.dump(best_params, f, indent=4)
    print(f" Saved best parameters to: {param_path}")

    # Evaluate model performance
    y_pred_train = best_model.predict(X_train)
    y_pred_test = best_model.predict(X_test)

    train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
    test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)
    test_r2 = r2_score(y_test, y_pred_test)

    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test  RMSE: {test_rmse:.4f}")
    print(f"Test   R : {test_r2:.4f}")

print("\n All models trained successfully.")
```

```python
# Set global random seed for reproducibility
SEED = 42
np.random.seed(SEED)

# Ensure output directories exist
os.makedirs("../output/models_with_city", exist_ok=True)
os.makedirs("../output/param_logs_with_city", exist_ok=True)

# One-hot encode the 'City' column
city_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
city_onehot = city_encoder.fit_transform(grid500[['City']])
city_feature_names = city_encoder.get_feature_names_out(['City'])

# Create one-hot DataFrame and concatenate
df_city_encoded = pd.DataFrame(city_onehot, columns=city_feature_names, index=grid500.index)
grid500_with_city = pd.concat([grid500, df_city_encoded], axis=1)

# ========================= XGBoost Training Loop =========================
targets = ['log_NTL2023_focal', 'log_VIT202311', 'UHIDAY2020_07_inv']

# Combine city features with existing morphological features
features_extended = features + list(city_feature_names)

for target in targets:
    print(f"\n===== Training model for target: {target} =====")

    # Drop rows with missing target or features
    df_target = grid500_with_city.dropna(subset=features_extended + [target])

    # Split features and target
    X = df_target[features_extended]
    y = df_target[target]

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=SEED
    )

    print(f"Training samples: {X_train.shape[0]}, Testing samples: {X_test.shape[0]}")

    # Define base XGBoost model
    xgb_model = xgb.XGBRegressor(
        objective='reg:squarederror',
        random_state=SEED,
        n_jobs=-1
    )

    # Randomized hyperparameter search
    param_dist = {
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [3, 5, 7, 9],
        'learning_rate': [0.01, 0.03, 0.05, 0.07, 0.1],
        'subsample': [0.7, 0.9, 1.0],
        'colsample_bytree': [0.7, 0.9, 1.0],
        'gamma': [0, 0.1, 0.3],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2],
        'min_child_weight': [1, 3, 5],
    }

    random_search = RandomizedSearchCV(
        estimator=xgb_model,
        param_distributions=param_dist,
        n_iter=200,
        cv=5,
        scoring='neg_mean_squared_error',
        verbose=1,
        random_state=SEED
    )

    # Fit model
    random_search.fit(X_train, y_train)

    # Retrieve best estimator
    best_model = random_search.best_estimator_

    # Save trained model
    model_path = f"../output/models_with_city/{target}_xgb_model.pkl"
    joblib.dump(best_model, model_path)
    print(f" Saved best model to: {model_path}")

    # Save best hyperparameters
    best_params = random_search.best_params_
    param_path = f"../output/param_logs_with_city/{target}_best_params.json"
    with open(param_path, 'w') as f:
        json.dump(best_params, f, indent=4)
    print(f" Saved best parameters to: {param_path}")

    # Evaluate model performance
    y_pred_train = best_model.predict(X_train)
    y_pred_test = best_model.predict(X_test)

    train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
    test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)
    test_r2 = r2_score(y_test, y_pred_test)

    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test  RMSE: {test_rmse:.4f}")
    print(f"Test   R : {test_r2:.4f}")

print("\n All models trained successfully with city one-hot encoding.")
```

- City dummy variables act as latent factors for unobservable systemic differences.
